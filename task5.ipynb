```python
import matplotlib.pyplot as plt
import numpy as np
print("Graph 1: Line Chart - y = 3x + 2")
x = np.arange(0, 11)
y = 3 * x + 2
plt.plot(x, y, color='blue', linewidth=2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("y = 3x + 2")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
print("Graph 2: Scatter Plot with Customization")
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, color='red', s=100)
plt.grid(True)
plt.title("Customized Scatter Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
print("Graph 3: Bar Chart with Annotations")
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 40, 35, 50, 20]
plt.bar(categories, values, color='skyblue')
for i, val in enumerate(values):
    plt.text(i, val + 1, str(val), ha='center')
plt.title("Bar Chart with Annotations")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
print("Graph 4: Multiple Lines in One Plot (sin, cos, exp)")
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), 'r-', label='sin(x)')
plt.plot(x, np.cos(x), 'g--', label='cos(x)')
plt.plot(x, np.exp(x/5), 'b:', label='exp(x/5)')
plt.legend()
plt.title("Multiple Lines in One Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
print("Graph 5: Histogram with Custom Bins")
data = np.random.randn(1000)
plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram with Custom Bins")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
print("Graph 6: Subplots Grid (Line, Bar, Scatter, Histogram)")
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
x = np.linspace(0, 10, 100)
axs[0, 0].plot(x, 3*x+2)
axs[0, 0].set_title("Line Plot")
axs[0, 1].bar(['A', 'B', 'C'], [5, 7, 3])
axs[0, 1].set_title("Bar Chart")
axs[1, 0].scatter(np.random.rand(10), np.random.rand(10))
axs[1, 0].set_title("Scatter Plot")
axs[1, 1].hist(np.random.randn(100))
axs[1, 1].set_title("Histogram")
fig.suptitle("2x2 Grid of Plots")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
print("Graph 7: Stack Plot - Cumulative Sales of 3 Products")
months = np.arange(1, 13)
product1 = np.random.randint(20, 50, size=12)
product2 = np.random.randint(10, 40, size=12)
product3 = np.random.randint(5, 30, size=12)
plt.stackplot(months, product1, product2, product3,
              labels=['Product 1', 'Product 2', 'Product 3'])
plt.legend()
plt.title("Cumulative Sales Over 12 Months")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
print("Graph 8: Boxplot Using DataFrame - Student Exam Scores")
data = {
    'Math': np.random.randint(50, 100, size=20),
    'Science': np.random.randint(50, 100, size=20),
    'English': np.random.randint(50, 100, size=20)
}
df = pd.DataFrame(data)
df.boxplot()
plt.title("Student Exam Scores Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt
print("Graph 9: Pie Chart with Explode")
companies = ['A', 'B', 'C', 'D']
market_share = [35, 25, 25, 15]
explode = [0.1 if val == max(market_share) else 0 for val in market_share]
plt.pie(market_share, labels=companies, autopct='%1.1f%%',
        explode=explode, shadow=True)
plt.title("Market Share of Companies")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
print("Graph 10: Axis and Legend Customization")
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label="sin(x)")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fontsize=12)
plt.gca().invert_xaxis()
plt.title("Axis and Legend Customization")
plt.show()
```
