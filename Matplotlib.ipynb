```python
import matplotlib.pyplot as plt
import matplotlib
```

```python
students = ["A", "B", "C", "D", "E"]
marks = [76, 88, 90, 66, 80]
```

```python
styles = matplotlib.style.available[:10]
```

```python
for style in styles:
    plt.style.use(style)
    plt.figure(figsize=(6,4))
    plt.bar(students, marks)
    plt.title(f"Style: {style}")
    plt.xlabel("Students")
    plt.ylabel("Marks")
    plt.ylim(0, 100)
    plt.show()
```
