```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = ['2001', '2003', '2005', '2007', '2009']
sachin_runs = [1200, 1100, 980, 1050, 1005]
kohli_runs = [0, 0, 500, 900, 1150]
sehwag_runs = [700, 850, 1000, 950, 880]
```

```python
x = np.arange(len(years))
width = 0.25
```

```python
plt.figure(figsize=(10,6))
plt.bar(x - width, sachin_runs, width, label='Sachin')
plt.bar(x, kohli_runs, width, label='Kohli')
plt.bar(x + width, sehwag_runs, width, label='Sehwag')
plt.xlabel('Years')
plt.ylabel('Runs')
plt.title('Runs by Sachin, Kohli, and Sehwag over Years')
plt.xticks(x, years)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
#horizontal bar chart of debut 5 years total runs
players = ['Sachin', 'Kohli', 'Sehwag']
debut_runs = [5000, 4800, 4200]
```

```python
plt.figure(figsize=(8,5))
plt.barh(players, debut_runs, color=['orange', 'green', 'blue'])
plt.xlabel('Total Runs')
plt.title('Total Runs in Debut 5 Years')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```

```python
labels = ['Android', 'iOS', 'Others']
sizes = [72, 26, 2]
colors = ['green', 'gray', 'lightblue']
```

```python
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Mobile OS Market Share - Pie Chart')
```

```python
plt.figure(figsize=(8,4))
plt.barh(labels, sizes, color=colors)
plt.title('Mobile OS Market Share - Bar Chart')
plt.xlabel('Market Share (%)')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```
